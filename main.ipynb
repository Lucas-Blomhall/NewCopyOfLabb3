21. ### Använd .apply() 
1. Undersök vilka kategorier det finns i kolumnen "state"
2. Skapa en funktion som baserat på värdena i kolumnen state är, modifierar kolumnen "usd_funding_goal" med en * om det står success, "!" för failure eller canceled, "---" för "suspended".
3. Displaya ett sample av 20 rader från din modifierade kolumn  

```python
import pandas as pd
df = pd.read_csv("kickstarter.csv")

unique_states = df['state'].unique()
print(unique_states)

# Här gör jag en if state där jag kollar om den är success eller inte. Jag vet att jag inte skulle använda if states men det känns rätt eftersom i uppgiften
# så står det så här "Skapa en funktion som baserat på värdena i kolumnen state är" och jag tycker att det blir enklare att läsa. Snälla säg till om jag behöver göra om.
def modify_funding_goal(row):
    if row['state'] == 'successful':
        return str(row['usd_funding_goal']) + '*'
    elif row['state'] in ['failed', 'canceled']:
        return str(row['usd_funding_goal']) + '!'
    elif row['state'] == 'suspended': 
        return str(row['usd_funding_goal']) + '---'
    else:
        return row['usd_funding_goal']

df['usd_funding_goal'] = df.apply(modify_funding_goal, axis=1)

df
```
